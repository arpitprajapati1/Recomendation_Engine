```python
import numpy as np
import pandas as pd 
#rating = pd.read_csv("ratings.csv")
#rating.head()
rating_popular_movie = pd.read_csv (r'rating_popular_movie.csv')
rating_popular_movie.head()
```

```python
rating_popular_movie.shape
```

```python
movie_features_df=rating_popular_movie.pivot_table(index='title',columns='userId',values='rating').fillna(0)
movie_features_df.head(4)
```

```python
movie_features_df.shape
```

```python
from scipy.sparse import csr_matrix

movie_features_df_matrix = csr_matrix(movie_features_df.values)
# form here it will act as a numpy array
from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movie_features_df_matrix)
```

```python
#query_index = np.random.choice(movie_features_df.shape[0])
query_index = np.random.choice(movie_features_df.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(movie_features_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)# gives the wholerow of query index
```

```python
movie_features_df.head(304)
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(movie_features_df.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, movie_features_df.index[indices.flatten()[i]], distances.flatten()[i]))
```
